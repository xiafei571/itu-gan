# batch_process

```python
import pandas as pd
```

```python
df_generator = pd.read_csv('./data/generated_data.csv')
df_generator = df_generator.drop(columns = ['Unnamed: 0'])
print(df_generator.shape)
# df_generator.head()
```

```python
columns = df_generator.columns
print(columns)
```

```python
print(folder_dict[columns[0][0]])
```

```python
def getColumnsDict(columns):
    column_dict = {}
    for column in columns:
        file_type = folder_dict[column[0]]
        path_list = column_dict.get(file_type, [])
        path_list.append(column)
        column_dict[file_type] = path_list
    
    return column_dict
```

```python
column_dict = getColumnsDict(columns)
print(column_dict.keys())
```

```python
column_dict['virtual']
```

```python
f = open("./type3time_processed.txt","r")
file_list = f.readlines()
# file_list[0][:19]
```

```python
ori_file_count = 10
```

```python
for idx in range(df_generator.shape[0]):
    curr_idx = idx % ori_file_count
    file_name = file_list[curr_idx]
    
    for file_type, file_columns in column_dict.items():
        json_object = getJsonObject('./data/itu-type3/'+file_type+'/'+file_name[:19])
        for column in file_columns:
            new_obj = change_value(json_object, column[3:], int(df_generator.iloc[idx][column]))
            writeToJson(file_path = './data/itu-type3-generated/'+file_type+'/'+file_name[:12]+str(idx)+'.json', json_data = new_obj)
            
```
